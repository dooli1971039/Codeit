### numpy array와 python list는 둘 모두 값을 저장한다는 점에서 비슷하다. 하지만 문법이 다르다.



더불어 python list에서 뺄셈,나눗셈은 오류가 난다.


성능차이도 있다. 이는 저장 방식의 차이가 있기 때문이다.
python list는 다양한 자료형을 넣을 수 있지만 numpy array는 같은 자료형을 넣을 수 있다.

python list는 값을 추가하고 제거하는 일에 주로 쓰고,
numpy array는 수치계산이 많고 복잡할 때, 행렬같이 다차원 배열이 많을 때 사용한다.

```python
import numpy as np

array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.max()) # 최댓값
print(array1.min()) # 최솟값
```

```python
array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.mean()) # 평균값
```

```python
array1 = np.array([8, 12, 9, 15, 16])
array2 = np.array([14, 6, 13, 21, 23, 31, 9, 5])


#특이하게 median은 numpy array의 메소드가 아니라 numpy의 메소드입니다.
#그래서 np.mediaon(~~) 이렇게 쓴다.  ~~.median() 이 아니라...

print(np.median(array1)) # 중앙값
print(np.median(array2)) # 중앙값
```

```python
array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.std()) # 표준 편차
print(array1.var()) # 분산
```
